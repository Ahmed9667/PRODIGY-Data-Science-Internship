```python
# let us choose a country and show the population over years
country = df[df['Country Name'] == 'Egypt, Arab Rep.'] #that is our data frame

years = country.columns[1:]  #column 1960' index is 1
population = country.iloc[: , 1:].values.flatten() # values.flatten() to convert into values

plt.figure(figsize=(13,10))
sns.barplot(x=years , y= population)
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Population of Egypt per years')
plt.xticks(rotation = 90)
plt.show()

```

```python
# We can make a function to get the population of any country

def country(x):
    
    # let us choose a country and show the population over years
    country = df[df['Country Name'] == x] #that is our data frame

    years = country.columns[1:]  #column 1960' index is 1
    population = country.iloc[: , 1:].values.flatten()

    plt.figure(figsize=(13,10))
    sns.barplot(x=years , y= population)
    plt.xlabel('Years')
    plt.ylabel('Population')
    plt.title('Population of {} per years'.format(x))
    plt.xticks(rotation = 90)
    plt.show()
    
country('Venezuela, RB')
```
